```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Function to calculate Cramér's V
def cramers_v(confusion_matrix):
    chi2, p, dof, expected = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()  # total number of observations
    return np.sqrt(chi2 / (n * min(confusion_matrix.shape) - 1))



# Define the dependent variable
target = 'IncidentGrade'

# Independent variables
independent_vars = ['IncidentId', 'AlertId', 'DetectorId', 'AlertTitle', 'Category',
       'IncidentGrade', 'EntityType', 'EvidenceRole', 'DeviceId', 'Sha256',
       'IpAddress', 'Url', 'AccountSid', 'AccountUpn', 'AccountObjectId',
       'AccountName', 'DeviceName', 'NetworkMessageId', 'RegistryKey',
       'RegistryValueName', 'RegistryValueData', 'ApplicationId',
       'ApplicationName', 'OAuthApplicationId', 'FileName', 'FolderPath',
       'ResourceIdName', 'OSFamily', 'OSVersion', 'CountryCode', 'State',
       'City', 'Time']

# Calculate Cramér's V for each independent variable with respect to the target variable
results = {}
for var in independent_vars:
    # Create the confusion matrix for each independent variable vs target
    confusion_matrix = pd.crosstab(df[var], df[target])
    cramer_v_value = cramers_v(confusion_matrix)
    results[var] = cramer_v_value

# Display the results
for var, value in results.items():
    print(f"Cramér's V between {var} and {target}: {value:.3f}")

```

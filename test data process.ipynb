```python
df = df.astype({col: 'int32' for col in df.select_dtypes('int64').columns})# converting the int64 to int32 for insufficent storage
```

```python
df.isnull().sum()#finding the null values
```

```python
threshold= len(df)* 0.5   # removing columns with 50% above null value
df=df.dropna(thresh=threshold,axis=1)
```

```python
df.isnull().sum()
```

```python
#converting the timestamp columns to day, date, hour,year and time
df['Timestamp']=pd.to_datetime(df['Timestamp'])
df["Day"]=df["Timestamp"].dt.day
df["Month"]=df["Timestamp"].dt.month
df["Year"]=df["Timestamp"].dt.year
df["Hour"]=df["Timestamp"].dt.hour
df["Time"]=df["Timestamp"].dt.time
```

```python
df.drop(columns=['Timestamp','Usage'],axis=1,inplace=True)#droping the timestamp column and the column which was not present in training data
```
